```python
import pandas as pd
import numpy as np
```

```python
file1 =("pandashomeworkfolder/purchase_data.csv")
```

```python
purchasedata = pd.read_csv(file1)
```

```python
#player count
players = len(purchasedata)
numberofplayers = pd.DataFrame([players], columns = ["Total Players"])
#numberofplayers = pd.df([players]) = ["Total Players"]
numberofplayers
```

```python
#purchase analysis
averagepurchaseprice = purchasedata["Price"].mean()
averagepurchaseprice

averageprice1 = np.round(averagepurchaseprice,2)
averageprice1
```

```python
numberofpurchases = purchasedata["Price"].count()
numberofpurchases
```

```python
#uniqueitems = numberofpurchases
uniqueitems = purchasedata["Item ID"].nunique()
uniqueitems
```

```python
totalrevenue = purchasedata["Price"].sum()
totalrevenue
```

```python
purchasetable = pd.DataFrame({"Number Of Unique Items": [uniqueitems],
                             "Average Price": [averagepurchaseprice],
                             "Number Of Purchases": [numberofpurchases],
                             "Total Revenue": [totalrevenue]})
purchasetable.style.format({"Average Price":"${:.2f}", "Total Revenue":"${:.2f}"})
```

```python
#begin gender
genderdemo = purchasedata.groupby(["Gender"])
```

```python
purchasecount = genderdemo ["Gender"].count()
```

```python
totalcount = genderdemo ["Gender"].count()
totalcount1 = [purchasecount[0], purchasecount[1], purchasecount[2]]
```

```python
genderpercentage = [round((purchasecount[0]/numberofplayers)*100,2),round((purchasecount[1]/numberofplayers)*100,2),round((purchasecount[2]/numberofplayers)*100,2)]
```

```python
genderdemo = pd.DataFrame({"Percentage Of Players": genderpercentage, "Total Count": totalcount1})
genderdemo.index = (["Female Players", "Male Players", "Other/Non Disclosed"])
genderdemo
```

```python
genderdemo = purchasedata.groupby(["Gender"])
```

```python
#nextgender - purchasing analysis
purchasecount = genderdemo["SN"].count()
```

```python
purchaseprice = genderdemo["Price"].mean()
```

```python
purchasevalue = genderdemo["Price"].sum()
```

```python
duplicates = purchasedata.drop_duplicates(subset = "SN", keep = "first")
duplicates1 = duplicates.groupby(["Gender"])
```

```python
purchasebygender = (genderdemo["Price"].sum()/duplicates1["SN"].count())
```

```python
genderanalysis = pd.DataFrame({"Purchase Count": purchasecount,
                              "Average Purchase Price": purchaseprice,
                               "Total Purchase Value": purchasevalue,
                               "Avg Purchase By Gender": purchasebygender})
```

```python
#genderanalysis["Purchase Count"] = genderanalysis["Purchase Count"].map()
genderanalysis["Average Purchase Price"] = genderanalysis["Average Purchase Price"].map("${:.2f}".format)
genderanalysis["Total Purchase Value"] = genderanalysis["Total Purchase Value"].map("${:.2f}".format)
genderanalysis["Avg Purchase By Gender"] = genderanalysis["Avg Purchase By Gender"].map("${:.2f}".format)
genderanalysis = genderanalysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Purchase By Gender"]]
genderanalysis
```
